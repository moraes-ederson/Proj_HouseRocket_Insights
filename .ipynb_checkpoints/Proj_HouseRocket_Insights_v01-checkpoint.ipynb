## 1. Problema de Negócio

Trata-se de um problema fictício que visa encontrar as melhores oportunidades de compra e venda de imóveis disponíveis no portfólio da House Rocket, empresa também fictícia.

E quais são as dores enfrentadas pelos tomadores de decisão?  
- O time do negócio não consegue tomar boas decisões de compra e venda sem analisar os dados.  
- O portfólio é muito grande e leva muito tempo para fazer o trabalho manualmente.  

### 1.1 Objetivo

Gerar insights através da análise e manipulação dos dados para auxiliar na tomada de decisão pelo time de negócios.

### 1.2 Questões de negócios a serem respondidas

1 - Quais são os imóveis que a House Rocket deveria comprar e por qual preço?

2 - Uma vez o imóvel comprado, qual o melhor momento para vendê-lo e por qual preço?

### 1.3 Insights

Visando descobrir e entregar mais insights para o time de negócios, vamos explorar e validar as seguintes hipóteses:

- H1: Imóveis que possuem vista para água são pelo menos 30% mais caros, na média.
- H2: Imóveis com ano de construção menor que 1955 são pelo menos 50% mais baratos, na média.
- H3: Imóveis sem porão possuem terrenos pelo menos 50% maiores do que com porão, na média.
- H4: O crescimento do preço dos imóveis YoY (Year over Year) é de pelo menos 10%.
- H5: Imóveis com 3 banheiros tem um crescimento de preço MoM (Month over Month) de pelo menos 15%.

## 6. Respondendo as Questões de Negócios

O objetivo na sequência é preparar os dados para responder as questões:


**1. Quais são os imóveis que a House Rocket deveria comprar e por qual preço?**

Plano de execução:
- Agrupar os imóveis por região (zipcode).
- Dentro de cada região, eu vou encontrar a mediana do preço do imóvel.
- Vou sugerir que os imóveis abaixo do preço mediano da região e que estejam em boas condições sejam comprados.

```python
# Carregando novamente e visualizando os dados originais
df = pd.read_csv('dataset/kc_house_data.csv', parse_dates=['date'])
```

```python
# Criando coluna de preço mediano por zipcode e unindo ao dataset
df_median = df[['zipcode', 'price']].groupby('zipcode').median().reset_index()
df_median
```

```python
df2 = pd.merge(df, df_median, on='zipcode', how='inner')
df2.head()
```

```python
# Renomeando as colunas price_x e price_y geradas pelo merge
cols = ['price_x', 'price_y']
ren = ['price', 'median_price_zip']
df2.rename({cols[0]: ren[0], cols[1]: ren[1]}, axis=1, inplace=True)
```

A variável "condition" contém uma classificação de 1 a 5 sobre o estado (condição) do imóvel, sendo 1 para condição muito ruim e 5 para condição muito boa.

Portanto vamos considerar uma compra viável se o preço da casa estiver abaixo do preço mediano da região e com a classificação de condição maior ou igual a 3.

```python
for i in range(len(df2)):
    if (df2.loc[i, 'price'] < df2.loc[i, 'median_price_zip']) & (df2.loc[i, 'condition'] >= 3):
        df2.loc[i, 'recommendation'] = 'buy'
    else:
        df2.loc[i, 'recommendation'] = 'not buy'
```

```python
# Visualizando uma tabela com a recomendação de compra ou não.

data_report1 = df2[['id', 'zipcode', 'price', 'median_price_zip', 'condition', 'recommendation']].copy()

# Adicionando coluna para indicar o valor de economia no preço
data_report1['saving'] = data_report1['median_price_zip'] - data_report1['price']

data_report1.to_csv('reports/buy_recomendation.csv')
data_report1
```

```python
# Quantidade dos imóveis recomendados para compra
data_report1['recommendation'].value_counts()
```

Vamos estimar quanto a empresa estaria economizando considerando a compra dos imóveis recomendados.  
Para isso vamos agrupar o valor de economia com a recomendação para compra e depois somar os valores.

```python
only_buy = data_report1.loc[data_report1['recommendation'] == 'buy']
total_save = only_buy['saving'].sum()
total_price = only_buy['price'].sum()
save_percentage = total_save / total_price * 100
print(f'A empresa tem potencial para economizar cerca de US$ {total_save:.2f} ({save_percentage:.2f}%) adquirindo os \
imóveis recomendados considerando valor mediano de mercado.')
```

**2. Uma vez a casa comprada, qual o melhor momento para vendê-las e por qual preço?**

Plano de execução:

- Agrupar os imóveis recomendados para compra por região (zipcode) e por sazonalidade (estações do ano).
- Dentro de cada região e sazonalidade, eu vou calcular a mediana do preço.
- Condições de venda:
    1. Se o preço da compra for maior que a mediana da região + sazonalidade:
        O preço da venda será igual ao preço da compra + 10%.
    2. Se o preço da compra for menor que a mediana da região + sazonalidade.
        O preço da venda será igual ao preço da compra + 30%.

```python
# Selecionando apenas os imóveis recomendados para compra
df3 = df2.loc[df2['recommendation'] == 'buy'].copy()
df3.head()
```

Para entender a relação das estações do ano com o preço de venda do imóvel, vamos considerar a data do registro do imóvel no conjunto de dados.

Ao agrupar os imóveis por região e estação do ano em que foi anunciada, será possível comparar os preços medianos da região e por sazonalidade.

Por se tratar de imóveis a venda nos EUA, as estações do ano acontecem nas seguintes datas:
- **Inverno:** 21 de dezembro a 20 de março;
- **Primavera:** 21 de março a 20 de junho;
- **Verão:** 21 de junho a 22 de setembro;
- **Outono:** 23 de setembro a 20 de dezembro.

Para efeito de análise vou considerar as estações separadas por meses, logo:
- **Inverno:** jan a mar;
- **Primavera:** abr a jun;
- **Verão:** jul a set;
- **Outono:** out a dez.

```python
# Adicionando uma coluna de mês para auxiliar na criação da coluna de estações do ano
df3.insert(2, 'month', df3['date'].dt.month)
```

```python
# Criando a variável com as estações do ano
df3['season'] = df3['month'].apply(lambda x: 
                                   'winter' if 1 <= x <= 3 else 
                                   'spring' if 4 <= x <= 6 else 
                                   'summer' if 7 <= x <= 9 else 
                                   'fall')
```

```python
# Agrupando os preços medianos por região e estações do ano
df3_median_season = df3[['price', 'zipcode', 'season']].groupby(['zipcode', 'season']).median().reset_index()
df3_median_season
```

```python
# Renomeando a coluna 'price'
df3_median_season.rename({'price': 'median_price_season'}, axis=1, inplace=True)
```

```python
# Unindo os valores de preços medianos ao dataset completo
df4 = pd.merge(df3, df3_median_season, on=['zipcode', 'season'])
df4.head(3)
```

```python
# Criando a coluna para recomendação de preço de venda
for i in range(len(df4)):
    if df4.loc[i, 'price'] > df4.loc[i, 'median_price_season']:
        df4.loc[i, 'sell_price'] = df4.loc[i, 'price'] * 1.1
    else:
        df4.loc[i, 'sell_price'] = df4.loc[i, 'price'] * 1.3
```

A coluna 'sell_price' mostra o valor recomendado de venda para cada imóvel considerando todos eles comprados no preço anunciado. O valor levou em consideração a região do imóvel e a estação do ano em que foi anunciado para comparar com o preço mediano dos imóveis por região e época do ano.

E por fim, vou adicionar a coluna 'profit' para visualizar o possível lucro caso a venda ocorra no valor recomendado.

```python
# Adicionando a variável 'profit' para apresentar o lucro que a empresa espera atingir com as recomendações de venda
df4['profit'] = df4['sell_price'] - df4['price']
```

```python
# Gerando o relatório em arquivo csv para o time de negócios
data_report2 = df4[['id', 'zipcode', 'season', 'price', 'median_price_season', 'sell_price', 'profit']].copy()
data_report2.to_csv('reports/sale_recomendation.csv')
data_report2
```

```python
# Verificando preço médio em relação a estação do ano
data_report2[['season', 'price']].groupby('season').mean().sort_values('price', ascending=False)
```

A próxima etapa é checar o potencial lucro bruto estimado da empresa considerando as vendas dos imóveis com os valores recomendados.

```python
total_profit = data_report2['profit'].sum()
total_price = data_report2['price'].sum()
percentage_profit = total_price / total_profit * 100
print(f'A empresa tem potencial para lucrar cerca de US$ {total_profit:.0f} ({percentage_profit:.0f}%) considerando a \
venda dos imóveis adquiridos nos valores recomendados.')
```

## 7. Análise Exploratória e Validação das Hipóteses

- H1: Imóveis que possuem vista para água são pelo menos 30% mais caros, na média.
- H2: Imóveis com ano de construção menor que 1955 são pelo menos 50% mais baratos, na média.
- H3: Imóveis sem porão possuem terrenos pelo menos 50% maiores do que com porão, na média.
- H4: O crescimento do preço dos imóveis YoY (Year over Year) é de pelo menos 10%.
- H5: Imóveis com 3 banheiros tem um crescimento de preço MoM (Month over Month) de pelo menos 15%.

```python
# Carregando novamente e visualizando os dados originais
df = pd.read_csv('dataset/kc_house_data.csv', parse_dates=['date'])
df.head()
```

### H1: Imóveis que possuem vista para água são pelo menos 30% mais caros, na média.

```python
# Agrupando o preço médio por imóveis com ou sem vista para água
df_H1 = round(df[['price', 'waterfront']].groupby('waterfront').mean().reset_index(), 2)
df_H1.columns = ['waterfront', 'mean_price']
df_H1['waterfront'] = df_H1['waterfront'].apply(lambda x: 'no' if x == 0 else 'yes')
df_H1
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x='waterfront', y='mean_price', data=df_H1)
plt.xlabel('Com Vista p/ Água', fontsize = 10.0, weight = 'bold')
plt.ylabel('Preço Médio (US$)', fontsize = 10.0, weight = 'bold')
plt.title('HIPÓTESE 1', weight = 'bold');
```

```python
# Fazendo a comparação dos preços
no_front = df_H1.loc[0, 'mean_price']
yes_front = df_H1.loc[1, 'mean_price']
overbalance = round(((yes_front - no_front) / no_front) * 100)
print(f'{overbalance}%')
```

**H1 é verdadeira. Como visto, os imóveis com vista para água são muito mais caros, sendo 213% acima da média comparado aos imóveis que não possuem vista similar.**

### H2: Imóveis com ano de construção menor que 1955 são pelo menos 50% mais baratos, na média.

```python
# Gerando um dataset apenas com o preço e ano de construção
df_H2 = df[['price', 'yr_built']].copy()
```

```python
# Dividindo os dados em categorias de mais velhos e mais novos do que 1955 para encontrar a média de preço delas
df_H2['age_built'] = df_H2['yr_built'].apply(lambda x: 'older' if x <= 1955 else 'newer')
df_H2 = round(df_H2[['price', 'age_built']].groupby('age_built').mean().reset_index(), 2)
df_H2.columns = ['age_built', 'mean_price']
df_H2
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x='age_built', y='mean_price', data=df_H2)
plt.xlabel('Idade do Imóvel', fontsize = 10.0, weight = 'bold')
plt.ylabel('Preço Médio (US$)', fontsize = 10.0, weight = 'bold')
plt.title('HIPÓTESE 2', weight = 'bold');
```

```python
# Fazendo a comparação das médias dos preços
newer = df_H2.loc[0, 'mean_price']
older = df_H2.loc[1, 'mean_price']
overbalance = round(((newer - older) / newer) * 100)
print(f'{overbalance}%')
```

**H2 é falsa. Como visto, os imóveis mais antigos classificados com o ano de construção inferior a 1955 não são 50% mais baratos na média. Pelo contrário, a média de preços é praticamente a mesma independente do ano de construção do imóvel.**

### H3: Imóveis sem porão possuem terrenos pelo menos 50% maiores do que com porão, na média.

```python
# Gerando um dataset apenas com as variáveis de área do terreno e área do porão
df_H3 = df[['sqft_lot', 'sqft_basement']].copy()
```

```python
# Checando alguns valores aleatórios da variável da área do porão
df_H3['sqft_basement'].sample(10)
```

Vemos que há imóveis com área de porão igual a zero, ou seja, não possuem porão. Com isso, vou criar uma variável com a categoria "tem" porão (sim) e "não tem" porão (não).

```python
# Dividindo os dados em categorias sim e não para encontrar a média de área do terreno para imóveis com e sem porão
df_H3['basement'] = df_H3['sqft_basement'].apply(lambda x: 'no' if x == 0 else 'yes')
df_H3 = round(df_H3[['basement', 'sqft_lot']].groupby('basement').mean().reset_index(), 2)
df_H3.columns = ['basement', 'mean_sqft_lot']
df_H3
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x='basement', y='mean_sqft_lot', data=df_H3)
plt.xlabel('Com Porão', fontsize = 10.0, weight = 'bold')
plt.ylabel('Média da Área de Terreno (pés quadrados)', fontsize = 10.0, weight = 'bold')
plt.title('HIPÓTESE 3', weight = 'bold');
```

```python
no_basement = df_H3.loc[0, 'mean_sqft_lot']
yes_basement = df_H3.loc[1, 'mean_sqft_lot']
overbalance = round(((no_basement - yes_basement) / no_basement) * 100)
print(f'{overbalance}%')
```

**H3 é falsa. Como visto, os imóveis que não contam com porão possuem um terreno 18% maior em média, não atingindo 50%.**

### H4: O crescimento do preço dos imóveis YoY (Year over Year) é de pelo menos 10%, na média.

```python
# Gerando um dataset com as variáveis de ID, data de anúncio e preço
df_H4 = df[['id', 'date', 'price']].copy()
```

```python
# Checando os anos dos anúncios
df_H4['date'].dt.year.unique()
```

```python
# Calculando a média de preço por ano
df_H4 = round(df_H4[['date', 'price']].groupby(df_H4['date'].dt.year).mean().reset_index(), 2)
df_H4.columns = ['year', 'mean_price']
df_H4
```

```python
# Checando o crescimento da média ano após ano
df_H4['growth_YoY'] = df_H4['mean_price'].pct_change()
df_H4
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x='year', y='mean_price', data=df_H4)
plt.xlabel('Ano', fontsize = 10.0, weight = 'bold')
plt.ylabel('Preço Médio (US$)', fontsize = 10.0, weight = 'bold')
plt.title('HIPÓTESE 4', weight = 'bold');
```

**H4 é falsa. Como visto, o preço médio dos imóveis aumentou pouco mais de 0,5% apenas em relação ao ano anterior, não atingindo o crescimento de 10% ano a ano.**

### H5: Imóveis com 3 banheiros tem um crescimento de preço MoM (Month over Month) de pelo menos 15%.

```python
# Gerando um dataset com as variáveis de ID, data de anúncio, preço e qtde de banheiros
df_H5 = df[['id', 'date', 'price', 'bathrooms']].copy()
```

```python
# Selecionando apenas os imóveis com 3 banheiros
df_H5 = df_H5.loc[df_H5['bathrooms'] == 3].sort_values('date')
df_H5
```

```python
# Criando variável ano_mês para agrupar pelo preço médio de cada período
df_H5['year_month'] = df_H5['date'].apply(lambda x: 100*x.year + x.month)
df_H5['year_month']
```

```python
# Agrupando o preço médio por ano e mês
df_H5 = round(df_H5[['year_month', 'price']].groupby('year_month').mean().reset_index(), 2)
df_H5.columns = ['year_month', 'mean_price']
df_H5
```

```python
# Checando o crescimento da média mês após mês
df_H5['growth_MoM'] = df_H5['mean_price'].pct_change()
df_H5
```
